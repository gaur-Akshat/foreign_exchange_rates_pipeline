```python
import pandas as pd
from pathlib import Path

silver_file = Path("exchange_rates/exchange_rates.parquet")
df = pd.read_parquet(silver_file)
print(f"Loaded {len(df)} rows, {df.shape[1]} columns")
df.head(10)
```

```python
df.dtypes
```

```python
print("Earliest date:", df["rate_date"].min().date())
print("Latest date:  ", df["rate_date"].max().date())
print("Total unique dates:", df["rate_date"].nunique())
```

```python
print("Base currencies: ", df["base_currency"].unique())
print("Target currencies:", sorted(df["target_currency"].unique()))
```

```python
nulls = df.isnull().sum()
print("Null counts per column:")
print(nulls[nulls > 0] if nulls.any() else "No nulls found ✅")
```

```python
dupes = df.duplicated(subset=["rate_date", "base_currency", "target_currency"]).sum()
print(f"Duplicate rows: {dupes}" if dupes else "No duplicates found ✅")
```

```python
df.groupby("target_currency")["exchange_rate"].agg(["min", "max", "mean"]).round(4)
```

```python
latest = df[df["rate_date"] == df["rate_date"].max()]
print(f"Latest snapshot date: {df['rate_date'].max().date()}")
latest[["target_currency", "exchange_rate"]]
```
